# Approximating abs(x)

### This code is designed to approximate the function abs(x) using an function approximation.

### First import the libraries

```python
import numpy as np   # import numpy
import math as math        # import math
import matplotlib.pyplot as plt
```

### Define the function that approximates abs(x)

```python
def abs_approx(x,N):
    summation = np.zeros(x.size)
    k=1
    while k <= N:
        summation = summation + (np.cos((2*k-1)*x)) / (2*k-1)**2
        k = k + 1
    f = (math.pi/2) - (4.0/math.pi)*summation
    return f
```

### set up the variables and plot abs(x)

```python
n = 101      # number of points in the x axis
x = np.linspace(-math.pi,math.pi,101)
absx = np.abs(x)
plt.plot(x,absx,'k-')
leglist = []
leglist.append('y=|x|')
plt.xlabel('x')
plt.ylabel('y=|x|');
```

### Run the main loop and make the plot

```python
# plot abs(x)
plt.plot(x,absx,'k-')


N=1          # this is the number of wavelenths
Nmax = 4     # max value of N to consider
##leglist = []
while N <= Nmax:
    fval = abs_approx(x,N)
    # now plot the curves
    plt.plot(x,fval)
##    leg = f'N={N}'
    leglist.append(f'N={N}')
    N = N + 1

# now plot the curves
plt.xlabel('x')
plt.ylabel('y=|x|')
plt.legend(leglist)
###plt.legend(['using a loop','vectorized approach'])
plt.title('Approximating |x|')
plt.savefig('approx_abs.png')
plt.show()
```

### The approximation for abs(x) gets better as N increases.
